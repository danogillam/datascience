```python
from __future__ import division
import numpy as np
import scipy.io
from lab8_common import *
from numpy.random.mtrand import multinomial 
from pyqtgraph.Qt import QtGui, QtCore
import matplotlib.pylab as pylab
```

## Proposal distribution

```python
#create a proposal distribution
def ray_lik_gaus(particles,sensor_data,room_map):
    ray = cast_rays(particles,room_map)
    diff = ray-list(sensor_data)
    sol =  [np.exp(-15*np.dot(i,i)) for i in diff]
    return sol/sum(sol)

def ray_lik_lap(particles,sensor_data,room_map):
    ray = cast_rays(particles,room_map)
    diff = ray-list(sensor_data)
    sol =  [np.exp(-15*np.sum(np.abs(i))) for i in diff]
    return sol/sum(sol)
```

## Resampling

```python
def resample(particles,weights):
    #choose which particles' gaussians to sample from
    M,N = np.shape(particles)
    
    dists = np.random.choice(np.arange(N),p=weights,size=N) 
    new_samp = np.array([particles[:,i] for i in dists]).T
    
    #add noise to new_samp
    new_samp[0] = [new_samp[0,c] + .1*np.random.rand()-.05           for c in xrange(N)]
    new_samp[1] = [new_samp[1,c] + .1*np.random.rand()-.05           for c in xrange(N)]
    new_samp[2] = [new_samp[2,c] + 11*np.random.rand()%(2*np.pi)     for c in xrange(N)]
    

    
    return new_samp
    
```

## Run iteration of Particle Filter

```python
def particle_filter(room_map,sensor_data,P,parts=10000,iters=0,movie=False):
    particles = np.random.rand(3,parts)
    if not iters: iters = np.shape(sensor_data)[1]
        
    particle_slides = []
    for i in xrange(iters):
        print '\r',i,'/',iters,
        weights = P(particles,sensor_data[:,i],room_map)
        particles = resample(particles,weights)
        particle_slides.append(particles)
    
    particle_means = np.mean(particle_slides,axis=2)
    
    if movie:
        %matplotlib 
        for i in xrange(iters):
            #play movie
            plt.figure( 1 )
            plt.clf()
            
            show_map(room_map)
            plt.scatter(true_states[0,:i],true_states[1,:i])
            plt.scatter(particle_slides[i][0],particle_slides[i][1],s=1,c='red')
            plt.plot(particle_means[:i,0],particle_means[:i,1],c='green')
            plt.title(i)
            
            plt.pause( 0.0001 )
        plt.close()
    else:
        %matplotlib inline
        pylab.rcParams['figure.figsize'] = 10, 10 
        plt.scatter(true_states[0,:i],true_states[1,:i],label='True State')
        plt.plot(particle_means[:,0],particle_means[:,1],c='green',label='Expected Value')
        plt.scatter(particle_slides[-1][0],particle_slides[-1][1],s=2,c='red',label='particles')
        show_map(room_map)
        plt.legend(bbox_to_anchor=(1.05,1),loc=2, fontsize = 'xx-large')
        plt.title('particle filter')
        plt.show()

room_map = create_map() 
tmp = scipy.io.loadmat('/home/danogillam/databases/particlefilter/sensors.mat')
sensor_data = tmp['sonars']  # frames of our video
true_states = tmp['true_states']  # our little template

particle_filter(room_map,sensor_data,ray_lik_gaus,movie=False)    
    
```
